```python
from zipfile import ZipFile
import pandas as pd

```

```python
zip = ZipFile("csv_365.zip") 
name=zip.namelist()
zip.extractall()
```

```python
l=pd.read_csv(name[0])
col=list(l.columns)[0].split(';')
col.insert(0,'Name')
```

```python
o=[]
for i in name:
    h=pd.read_csv(i)
    n=i.split('_')[0]
    for j in list(h.index):
        gg=h.iloc[j,0].split(';')
        gg.insert(0,n)
        o.append(gg)


    
 
```

```python
new_df=pd.DataFrame(o,columns=col)
```

```python
def Coloring(value):
    if value < 0:
        return "Red"
    if value > 0:
        return "Green"
    if value == 0:
        return 'Gray'
```

```python
new_df[['open', 'high', 'low', 'close', 'volume', 'marketCap']] = new_df[['open', 'high', 'low', 'close', 'volume', 'marketCap']].apply(pd.to_numeric)
new_df['DailyPrice'] = new_df['close']
new_df['CirculatingSupply'] = new_df['marketCap'] / new_df['DailyPrice']
new_df['Date'] = pd.to_datetime(new_df['timestamp'].str[1:11]).dt.date
new_df['DailyChange'] = (new_df['DailyPrice'] - new_df['DailyPrice'].shift(-1))
new_df['ChangeStatus'] = new_df['DailyChange'].map(Coloring)
new_df = new_df.drop(columns = 'timestamp')
```
